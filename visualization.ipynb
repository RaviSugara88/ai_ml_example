# Categorical vs Categorical

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset("tips")

# Check for categorical columns
print(df.select_dtypes(include=['category', 'object']).columns)
```

```python
sns.countplot(data=df, x='day', hue='sex')
plt.title('Count of Sex per Day')
plt.show()
```

```python
import pandas as pd

# Create proportion table
prop_df = df.groupby(['day', 'sex']).size().reset_index(name='count')
total_per_day = prop_df.groupby('day')['count'].transform('sum')
prop_df['proportion'] = prop_df['count'] / total_per_day

sns.barplot(data=prop_df, x='day', y='proportion', hue='sex')
plt.title('Proportion of Sex per Day')
plt.show()
```

```python
cross_tab = pd.crosstab(df['day'], df['sex'])

sns.heatmap(cross_tab, annot=True, fmt='d', cmap='Blues')
plt.title('Day vs Sex Heatmap')
plt.show()
```

```python
# FacetGrid with histplot using map_dataframe
g = sns.FacetGrid(df, col="species", height=4, aspect=1.2)
g.map_dataframe(sns.histplot, x="sepal length (cm)", bins=15, color="teal")
g.set_titles(col_template="Species {col_name}")
g.set_axis_labels("Sepal Length (cm)", "Count")
plt.suptitle("Sepal Length Distribution per Species", y=1.05)
plt.tight_layout()
plt.show()
```

## Boxplot for Outlier Detection

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris

# Load data
iris = load_iris(as_frame=True)
df = iris.frame
df["species"] = iris.target

# Drop non-numeric/categorical target column
df_melted = df.drop(columns='species').melt(var_name="Feature", value_name="Value")

# Plot
plt.figure(figsize=(10, 6))
sns.boxplot(x="Feature", y="Value", data=df_melted, palette="coolwarm")
plt.title("Boxplot for Outlier Detection")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Checking for Correlation (Multicollinearity)

```python
# Correlation matrix
corr = df.drop(columns="species").corr()
sns.heatmap(corr, annot=True, cmap="vlag", center=0)
plt.title("Correlation Heatmap")
plt.show()
```

# Checking Relationships

```python
# Scatter with hue
sns.scatterplot(data=df, x="sepal length (cm)", y="sepal width (cm)", hue="species", palette="Set1")
plt.title("Relationship - Sepal Length vs Sepal Width by Species")
plt.show()
```

# Checking Normal Distribution

```python
# KDE with normality check
for col in df.columns[:-1]:
    sns.kdeplot(df[col], fill=True, label=col)
plt.title("KDE Plot for Normality Check")
plt.legend()
plt.show()
```

```python
import scipy.stats as stats
import matplotlib.pyplot as plt

# Q-Q plot for Sepal Length
stats.probplot(df["sepal length (cm)"], dist="norm", plot=plt)
plt.title("Q-Q Plot - Sepal Length")
plt.show()
```

# Distribution of Target Variable (Imbalanced Classes)


```python
sns.countplot(x="species", data=df, palette="Set2")
plt.title("Class Distribution (Target Variable)")
plt.show()
```

# Feature Importance Visualization (After Model)

```python
from sklearn.ensemble import RandomForestClassifier

X = df.drop(columns='species')
y = df['species']
model = RandomForestClassifier()
model.fit(X, y)

# Feature importance
importance = model.feature_importances_
sns.barplot(x=X.columns, y=importance, palette="Spectral")
plt.title("Feature Importance (Random Forest)")
plt.ylabel("Importance Score")
plt.xticks(rotation=45)
plt.show()
```
